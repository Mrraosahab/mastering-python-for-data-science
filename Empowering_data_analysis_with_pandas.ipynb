# DataFrame

```python
# DataFrame is 2-dimension array
# DataFrame can be created from a dictionary of series
dict = {'c1':pd.Series(['a', 'b', 'c']), 'c2':pd.Series([10, 20, 30])}
df = pd.DataFrame(dict)
```

```python
df
```

```python
# DataFrame can be created using dictionary of two list too
dict1 ={'c1': ['A', 'B', 'C', 'D'], 'c2': [1, 2, 3, 4]}
df = pd.DataFrame(dict1)
```

```python
# filling missing values
# let's define a DataFrame
df = pd.DataFrame(np.random.randn(5,3), index=['a0', 'a10', 'a20', 'a30', 'a40'], columns=['X', 'Y', 'Z'])
```

```python
# print the new created dataFrmae
df
```

```python
# adding some more index to DataFrame these takes null value
df = df.reindex(['a0', 'a1', 'a10', 'a11', 'a20', 'a21', 'a30', 'a31', 'a40', 'a41'])
```

```python
# printing DataFrame
df
```

```python
# using fillna() we can fill any values in dataframe of our choice
# let's fill '0' replacing null values
df.fillna(0)
```

```python
# if we want to fill the values with mean of columns
df.fillna((df.mean()))
```
